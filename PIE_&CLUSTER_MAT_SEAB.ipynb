```python
 df = pd.read_excel('detail_transaksi.xlsx')
```

```python
# Use seaborn style for visual
sns.set(style="whitegrid")
```

```python
# filter rows from KOMISI = ST and Payment status = PAID
flt = df[(df['KOMISI']=='ST') & (df['PAYMENT_STATUS']=='PAID')]
```

```python
# assuming each row is one account, so count rows per city
# Group by city and count total accounts ( counting unique CUST_ACCT)
group_by_city = flt.groupby('CITY').size()
```

```python
# Calculate total accounts
Total_new_accounts = group_by_city.sum()
```

```python
# Plot pie chart
plt.figure(figsize=(10, 10))

# plot color pie chart
#cl = sns.color_palette(['#264653', '#E63946', '#F4A261', '#2A9D8F'])[:len(group_by_city)]
cl = sns.color_palette(['#264653','#2B5783', '#F4A261', '#4B4B4B', '#5C4033'])[:len(group_by_city)]
```

```python
# Define a function to show both percentage and absolute counts in label of pie chart
def autopct_format(p):
  absolute = int(round(p * Total_new_accounts /100.0))
  return f"{p:.0f}%\n({absolute})"
```

```python
plt.pie(
    group_by_city,
    labels=group_by_city.index,
    autopct=autopct_format,
    colors=cl,
    startangle=180,
    textprops={'fontsize': 12}
)
plt.title(f'Total_new_accounts by City\n: {Total_new_accounts}', fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
